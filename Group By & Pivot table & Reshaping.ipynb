```python
import pandas as pd
ratings = pd.read_csv(r"A:\Hari\ml-latest-small\ml-latest-small\ratings.csv")
ratings.head()
```

```python
ratings.groupby(['movieId'])['rating'].mean()
```

```python
ratings.groupby(['movieId']).rating.mean()
```

```python
ratings[['movieId','rating']].groupby('movieId').mean()
```

```python
ratings[['movieId','rating']].groupby('movieId').count()
```

```python
ratings.groupby(['movieId']).mean()
```

```python
ratings.groupby(['movieId']).mean()['rating']
```

```python
ratings.groupby(['movieId']).mean().reset_index()
```

```python
ratings.groupby(['movieId'],as_index=False).mean()
```

```python
ratings.userId.nunique()
```

```python
ratings.groupby(['userId'],as_index=False).mean()[['userId','rating']]
```

```python
# multiple aggregation is very useful we can do it pretty easily
ratings.groupby('movieId').rating.agg(['min','max','mean','count'])
```

# pivot table

```python
# pivot_table allows you to aggregate and summarize data
ratings.pivot_table(values='rating', index='movieId', columns=None, aggfunc='mean', margins=False)
# ratings[['movieId','rating']].groupby('movieId').mean()
# the above two will give the same result.
```

```python
ratings.pivot_table(values='rating', index='movieId', columns=None, aggfunc=['mean','count'], fill_value=None)
```

```python
ratings.pivot_table(values='rating', index='movieId', columns=['userId'], aggfunc='count', fill_value=None)
```
