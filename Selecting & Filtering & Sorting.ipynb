```python
import pandas as pd
movies = pd.read_csv("A:/Hari/imdbRatings.csv")
movies
```

# Selecting

```python
# "loc" method is used to select required rows and columns.
# Here we used slicing & fancy indexing inside loc.
# movies[ROW,COLUMN]
# while using loc the required rows and columns can be pointed using indexes and columns
movies.loc[:5,]
```

```python
movies.loc[:5,['title','star_rating']]
```

```python
# Here we used slicing & fancy indexing inside loc.
movies.loc[45:50,['title','star_rating']]
```

```python
# we can use both slicing and both fancy indexing as well
movies.loc[[1,46],['title','star_rating']]
```

```python
# here we used slicing for row and column
# inslicing we take the full range from lower to upper ie., 0:5 --> 0,1,2,3,4,5
movies.loc[:5,'star_rating':'genre']
```

```python
# by running the below code it will return a series.
movies.loc[4,]
```

```python
# "iloc" can also be used for selecting rows and columns
# The difference between "iloc" and "loc" is in "iloc" you don't have to specify the exact index and column name
# instead you have to specify row num and column num
movies.iloc[:5,:]
```

```python
# 0:5 it represents first five rows from 0 to 4 not 5.we have a upper bound -1 logic here [0,1,2,3,4]
# here '0' and '1' represents 'star_rating' and 'title'
movies.iloc[:5,[0,1]]
```

```python
first5  = movies.iloc[:5,[0,1]]
first5
```

```python
first5.index = ['a','b','c','d','e']
first5
```

```python
first5.iloc[:5,[0,1]]
```

```python
# that ix provides the usage of "loc" and "iloc"
first5.ix['a':'e',:]
```

```python
first5.ix[0:,[1,0]]
```

# Filtering

```python
# To get the data with certain condition we can use the code below
# there are many ways for conditional filtering  but using "loc" is the ideal way
movies.loc[movies["duration"]==movies.duration.max()]
```

```python
# both are same
movies[movies["duration"]==movies.duration.max()]
```

```python
# It can give the same result because it can perform "loc" operation
movies.ix[movies["duration"]==movies.duration.max()]
```

```python
movies.loc[movies["duration"] > 120]
```

```python
# The above code and this code gives same result
# but using the above one is more preferreble.
movies.loc[movies.duration > 120]
```

```python
# we can use muliple conditional filters in "loc" using "&" and "|" 
# while using multiple conditions seperate it with ()
movies.loc[(movies["duration"] > 120) & (movies["star_rating"] > 7.5)]
```

```python
movies.loc[(movies["duration"] > 120) & (movies["star_rating"] > 7.5)&(movies["genre"] == "Crime")]
```

```python
# pipelining is a very powerful thing in pandas we can execute series of operations with pipelining
movies.loc[movies.duration > 120].shape
```

```python
movies.loc[(movies["duration"] > 120) & (movies["star_rating"] > 7.5)].shape
```

```python
movies.loc[(movies["duration"] > 120) & (movies["star_rating"] > 7.5)&(movies["genre"] == "Crime")].shape
```

```python
movies.genre.unique()
```

```python
movies.loc[(movies["genre"] == "Crime") | (movies["genre"] == "Thriller")|(movies["genre"] == "Comedy")]
```

```python
movies.loc[movies.genre.str.contains(r'Crime|Thriller|Comedy')]
```

```python
movies.loc[movies.genre.isin(['Crime','Thriller','Comedy'])]
```
