```python
import pandas as pd
import numpy as np
movies = pd.read_csv("A:/Hari/imdbRatings.csv")
movies.head()
```

```python
# We can use python string functions using "str"
movies.actors_list = movies.actors_list.str.replace("[","").str.replace("]","")
movies.head()
```

```python
# To take a copy
moviescp = movies.copy()
moviescp.head()
```

```python
moviescp.columns
```

```python
# Here you have to give all columns name even you don't want to change
moviescp.columns = ["StarRating","title","contentRating","genre","duration","actorList"]
moviescp.head()
```

```python
moviescp = movies.copy()
moviescp.columns
```

```python
# to change specific columns
moviescp = moviescp.rename(columns = {'star_rating':'starRating','content_rating':'contentRating','actors_list':'actorList'})
moviescp.head()
```

```python
movies.head()
```

```python
# adding new columns
# you add a new colum based on the new column or a independent one
moviescp['Action'] = moviescp['genre'] == 'Action'
moviescp.head()
```

```python
moviescp["Excellent"] = moviescp['StarRating'] > 8
moviescp.head()
```

```python
# To insert columns at the desired location you can use insert
moviescp2 = movies.copy()
moviescp2.insert(4,"Action",moviescp2['genre'] == 'Action')
moviescp2.insert(5,"Excellent",moviescp2['star_rating']>8)
moviescp2.head()
```

```python
# you can create a new column based on the old column then give your custom values based on the condition using numpy
moviescp3 = movies.copy()
moviescp3.insert(4,"Action",np.where(moviescp2['genre'] == 'Action',1,0))
moviescp3.insert(5,"Excellent",np.where(moviescp2['star_rating']>8,"Yes","No"))
moviescp3.head()
```

```python
moviescp3.dtypes
```

```python
# There are many scenarios you need to type cast the fields
moviescp4 = moviescp3.copy()
moviescp4['Action']= moviescp4['Action'].astype("float")
moviescp4["duration"] = moviescp4["duration"].astype("float")
moviescp4.dtypes
```

```python
moviescp4.head(3)
```

```python
moviescp5 = moviescp3.copy()
moviescp5.head(3)
```

```python
# to change muliple column datatype in one line 
moviescp5 = moviescp5.astype({'Action':'float','duration':'float'})
moviescp5.head(3)
```
