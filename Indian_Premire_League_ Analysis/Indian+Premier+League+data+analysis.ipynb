```python
Batsman_score=delivery.groupby('batsman')["batsman_runs"].agg(sum).reset_index().sort_values(by="batsman_runs",ascending=False).reset_index(drop=True)
#Lets have a look what is happening 
# we group our deliveries data by Batsman ans then we sum the batsman_runs mean run scored by batsman
# Then sort those values in decreasing order
# reset_index means we are indexing again our data which changed due to first group by then by sort
Top_batsman_score= Batsman_score.iloc[:15,:] # here I am taking only top 10 scorer in IPL
Top_batsman_score # this will show our data
```

```python
#this is the another way of doing the same
delivery[['batsman',"batsman_runs"]].groupby('batsman').sum().sort_values(by ='batsman_runs',ascending = False)[:15]
```

```python
top_batsman_score = delivery[['batsman',"batsman_runs"]].groupby('batsman').sum().sort_values(by ='batsman_runs',ascending = False).reset_index()[:15]
```

```python
%matplotlib inline
delivery[['batsman',"batsman_runs"]].groupby('batsman').sum().sort_values(by ='batsman_runs',ascending = False)[:15].plot(kind = 'bar',figsize =(18,9))
plt.xlabel('players')
plt.ylabel('runs')
plt.title('Top run scorers in ipl')
```
