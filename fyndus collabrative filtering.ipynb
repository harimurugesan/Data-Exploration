```python
ad1 = advertisement.groupby(['flyer_user_id'],as_index=False).mean()[['flyer_user_id','rating']]
ad1.rename(columns= {'rating':'avg_user_rating'},inplace = True)
```

```python
advertisement = pd.merge(advertisement,ad1)
advertisement
```

```python
advertisement['adj_rat'] = advertisement['rating'] - advertisement['avg_user_rating']
advertisement
```

```python
#advertisement.reset_index(inplace=True)
#advertisement.drop('index',axis = 1,inplace=True)
ad_df = advertisement.pivot_table(index = 'flyer_user_id', columns ='ad_id', values = 'rating').fillna(0)
ad_df.head()
```

```python
R = ad_df.as_matrix()
user_ratings_mean = np.mean(R, axis = 1)
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
R_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 50)
```

```python
np.dot(U, sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = R_df.columns)
```
