# using mlextend

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
ap6.num_of_orders.unique()
```

```python
basket = ap6.groupby(['transaction_id', 'order_id'])['num_of_orders'].sum().unstack().reset_index().fillna(0).set_index('transaction_id')
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules = rules.astype({'antecedents':'str','consequents':'str'})
```

```python
rules.antecedents = rules.antecedents.str.replace('\D+',"")
rules.consequents = rules.consequents.str.replace('\D+',"")
rules
```

```python
m1 = ap5.merge(order_items,on='order_id')
m1
```

```python
m2= m1[['transaction_id','order_id','product_id']]
m2
```

```python
m2.order_id =m2.order_id.astype('str')
```

```python
m3 = m2[['order_id','product_id']].merge(rules,right_on = 'antecedents' ,left_on =  'order_id').drop(['order_id','antecedents'],axis = 1).rename(columns={'product_id':'product_id1'})
m3
```

```python
m4= m2[['order_id','product_id']].merge(m3,right_on = 'consequents' ,left_on =  'order_id').drop(['order_id','consequents'],axis = 1).rename(columns={'product_id':'product_id2'})
m4
```
