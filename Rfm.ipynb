```python
quantiles = rfm.quantile(q=[0.25,0.50,0.75])
quantiles
```

```python
d = quantiles.to_dict()
d
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
def FMScore(x,p,d):
     if x <= d[p][0.25]:
        return 1
     elif x <= d[p][0.50]:
        return 2
     elif x <= d[p][0.75]: 
        return 3
     else:
        return 4
```

```python
rfm['R_Quartile'] = rfm['recency'].apply(RScore, args=('recency',d,))
rfm['F_Quartile'] = rfm['Frequency'].apply(FMScore, args=('Frequency',d,))
rfm['M_Quartile'] = rfm['Monetary'].apply(FMScore, args=('Monetary',d,))
rfm
```

```python
rfm['FM_Quartile'] = (rfm['F_Quartile'] + rfm['M_Quartile']) / 2
rfm
```

```python
rfm['RFM_Score'] = rfm['R_Quartile'].map(str) + rfm['F_Quartile'].map(str) + rfm['M_Quartile'].map(str)
rfm
```

```python
rfm.sort_values('RFM_Score',ascending = False)
```
