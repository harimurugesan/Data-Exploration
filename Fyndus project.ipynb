```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
ap1 = order.groupby(['transaction_id','order_id'])['coupon_code'].sum()
#.unstack().fillna(0)
```

```python
ap2 = pd.DataFrame(ap1).reset_index()
```

```python
ap2.dtypes
```

```python
ap2.coupon_code = ap2.coupon_code.astype('str')
```

```python
ap2
```

```python
ap2.coupon_code = ap2.coupon_code.str.replace("\w{2,20}","1")
ap2
```

```python
ap3 = ap2[['transaction_id','order_id']]
```

```python
ap3
```

```python
ap4  = ap3.groupby('transaction_id')['order_id'].apply(list)
ap4 = pd.DataFrame(ap4).reset_index()
```

```python
ap4
```

```python
 mb = ap4.order_id.values
```

```python
newmb = []
for i in mb:
    newmb.append(list(map(str,i)))
```

```python
newmb
```

```python
len(newmb)
```

```python
from apyori import apriori
ar = list(apriori(newmb, min_support=0.00001, min_confidence=0.2, min_lift=2, min_length=2,max_length =2))
ar = pd.DataFrame(ar)
ar
```
